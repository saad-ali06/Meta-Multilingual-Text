```python
import re

def num_to_words(num):
    if num == 0:
        return "zero"

    under_20 = [
        'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 
        'eleven', 'twelve', 'thirteen', 'fourteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'nineteen'
    ]
    
    tens = [
        'twenty', 'thirty', 'forty', 'fifty', 'sixty', 'seventy', 'eighty', 'ninety'
    ]
    
    thousands = [
        'thousand', 'million', 'billion', 'trillion'
    ]
    
    def words(n):
        if n < 20:
            return under_20[n-1:n]
        elif n < 100:
            return [tens[(n // 10) - 2]] + words(n % 10)
        elif n < 1000:
            return [under_20[(n // 100) - 1]] + ['hundred'] + words(n % 100)
        else:
            for i, word in enumerate(thousands, 1):
                if n < 1000 ** (i + 1):
                    return words(n // 1000 ** i) + [word] + words(n % 1000 ** i)

    return ' '.join(words(num)).strip()

def replace_numbers_with_words(s):
    def replacer(match):
        num = int(match.group(0))
        return num_to_words(num)

    return re.sub(r'\b\d+\b', replacer, s)

# Test the function
test_string = "There are 2101 students."
converted_string = replace_numbers_with_words(test_string)
print(converted_string)  # Output: "There are twenty one students."

test_string2 = "I have 1000 dollars."
converted_string2 = replace_numbers_with_words(test_string2)
print(converted_string2)  # Output: "I have one thousand dollars."
```
